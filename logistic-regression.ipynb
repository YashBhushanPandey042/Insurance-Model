```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("/kaggle/input/age-subscription/age_subscription_dataset.csv")
df
```

```python
sns.scatterplot(x="Age",y="Subscription_Purchased",data=df)
plt.show()
```

```python
x=df[["Age"]]
y=df["Subscription_Purchased"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[4]])
```

```python
sns.scatterplot(x="Age",y="Subscription_Purchased",data=df)
sns.lineplot(x="Age",y=lr.predict(x),data=df,color="red")
plt.show()
```
