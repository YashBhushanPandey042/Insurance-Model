```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler 
```

```python
df=pd.read_csv("/kaggle/input/insurance/insurance.csv")
df
```

```python
en_data=df[["sex","smoker"]]
en_data
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop="first")
ar=ohe.fit_transform(en_data).toarray()
ar
```

```python
df_encoded=pd.DataFrame(ar,columns=["sex_male","smoker_yes"])
df_encoded
```

```python
df_numeric = df.drop(["sex", "smoker"], axis=1)
```

```python


# Concatenate encoded data
df = pd.concat(
    [df_numeric.reset_index(drop=True), df_encoded.reset_index(drop=True)], 
    axis=1
)
df
```

```python
x=df[["age","bmi","sex_male","smoker_yes","children"]]
y=df["charges"]
```

```python
# sns.pairplot(data=df)
# plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
